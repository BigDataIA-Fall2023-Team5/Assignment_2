```python
import requests
from bs4 import BeautifulSoup


url ="https://www.sec.gov/forms"
page = requests.get(url)
soup = BeautifulSoup(page.content,'html.parser')
result = soup.find()
number = result.find_all("td", class_= "release-number-content views-field views-field-field-release-number is-active")
description = result.find_all("td", class_= "display-title-content views-field views-field-field-display-title")
date = result.find_all("td", class_="noFilter views-field views-field-field-date")
sec_number = result.find_all("td", class_="list-page-detail-content views-field views-field-field-list-page-det-secarticle")
topic = result.find_all("td", class_="views-field views-field-term-node-tid")
```

```python
pdf_links = soup.select('a[href$=".pdf"]')

# Extract the href attributes
pdf_links2 = [link['href'] for link in pdf_links]
pdf_urls = []
# Print the extracted PDF URLs
for url in pdf_links2:
    url2='https://www.sec.gov'+url
    pdf_urls.append(url2)

print(pdf_urls)
```

```python
#for a list of values in the number header
number_list=[]
for result in number:
            number_list.append(result.text)

import re
numbers=[]
for i in number_list:
  match = re.search(r'\nNumber:(.+?)\n\s+', i)  # Use regular expression to match and extract the value
  if match:
    value = match.group(1)
    numbers.append(value)
  else:
    value = None
    numbers.append(value)
print(numbers)
```

```python
#for a list of values in the description header
description_list=[]
for result in description:
            description_list.append(result.text)

descriptions=[]
for i in description_list:
  match = re.search(r'\nDescription:\n(.+?)\n\n\n', i)  # Use regular expression to match and extract the value
  if match:
    value = match.group(1)
    descriptions.append(value)
  else:
    value = None
    descriptions.append(i)
print(descriptions)
```

```python
#for a list of values in the date header
date_list=[]
for result in date:
            date_list.append(result.text)

dates=[]
for i in date_list:
  match = re.search(r'\nLast Updated:(.+?)\n', i)  # Use regular expression to match and extract the value
  if match:
    value = match.group(1)
    dates.append(value)
  else:
    value = None
    dates.append(i)
print(dates)
```

```python
#for a list of values in the sec_number header
sec_number_list=[]
for result in sec_number:
            sec_number_list.append(result.text)

sec_numbers=[]
def extract_value(s):
    # Regular expression pattern to capture values after "SEC Number:"
    pattern = r'\nSEC Number:([^\n]*)\n'
    match = re.search(pattern, s)

    # If a match is found
    if match:
        value = match.group(1).strip()  # Extract and strip to remove leading/trailing spaces
        return value if value else None  # Return None if value is empty
    return None

for i in sec_number_list:
  sec_numbers.append(extract_value(i))
print(sec_numbers)
```

```python
#for a list of values in the topic header
topic_list=[]
for result in topic:
            topic_list.append(result.text)

topics=[]
def extract_value2(s):
    # Regular expression pattern to capture values after "SEC Number:"
    pattern = r'\nTopic\(s\):(.+?)\n\s+'
    match = re.search(pattern, s)

    # If a match is found
    if match:
        value = match.group(1).strip()  # Extract and strip to remove leading/trailing spaces
        return value if value else None  # Return None if value is empty
    return None
for i in topic_list:
  topics.append(extract_value2(i))
print(topics)
```

```python
import pandas as pd

# Sample data where each row represents a column of the DataFrame
data = [
    numbers,
    descriptions,
    dates,
    sec_numbers,
    topics,
    pdf_urls
]

# Convert data to DataFrame and then transpose
df_overall = pd.DataFrame(data).transpose()

# Optionally, set column names
df_overall.columns = ['Number', 'Description', 'Last_Updated','SEC_Number', 'Topic', 'PDF_Link']

print(df_overall)
```

```python
# Extracting using PyPDF

import requests
from pypdf import PdfReader
from io import BytesIO

def extract_pdf(link):
    response = requests.get(link)
    if response.status_code == 200:
        with BytesIO(response.content) as open_pdf_file:
            reader = PdfReader(open_pdf_file)
            text = ""
            for page_num in range(len(reader.pages)):
                text += reader.pages[page_num].extract_text()
    else:
        text = ""
    return text


for i in range(0,len(df_overall)):
    if df_overall.PDF_Link[i] != 'None':
        print(i)
        text = extract_pdf(df_overall.PDF_Link[i])
        words = re.sub(r'[^a-zA-Z0-9 ]', ' ', text)
        cleaned_text = ' '.join(words.split())
        df_overall['pyPDF_extraction'][i] = cleaned_text
    else:
        df_overall['pyPDF_extraction'] = ""
```

```python
df_overall['pyPDF_extraction'] = df_overall['pyPDF_extraction'].astype(str)
```

```python
def limit_string_length(text, max_tokens):
    tokens = text.split()
    truncated_tokens = tokens[:max_tokens]
    truncated_text = ' '.join(truncated_tokens)
    return truncated_text

# Set the maximum token limit
max_token_limit = 2000  # Adjust this to your desired limit

# Apply the function to the 'text_column' and create a new column with truncated text
df_overall['pyPDF_extraction'] = df_overall['pyPDF_extraction'].apply(lambda x: limit_string_length(x, max_token_limit))
```

```python
def count_tokens(text):
    tokens = text.split()
    return len(tokens)

df_overall['tokens'] = df_overall['pyPDF_extraction'].apply(lambda x: count_tokens(x))
```

```python
df = df_overall.dropna().reset_index().drop('index',axis=1)
```
